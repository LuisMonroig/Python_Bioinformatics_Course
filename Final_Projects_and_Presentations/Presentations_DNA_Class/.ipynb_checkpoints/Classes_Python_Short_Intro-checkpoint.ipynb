# Classes in Python


### Definition of a Python Class
A _Python Class_ is a programm that defines a custom made object by containing, in an organized manner, the attributes and methods of the object. Classes provide the structure for creating multiple objects of the same type.

Learning to use clases is essential for making projects that are more efficient, easier to read and easier to update than strictly procedural programming code. Classes also have the benefit of porviding a reciclable structure for defining diferent objects of the same type and their respective atributes. To undestand better the fundamental properties that define an object oriented programming language such as Python, consider watching the following video: [Object Oriented Porgramming in 7 minutes](https://www.youtube.com/watch?v=pTB0EiLXUC8).

In the process of creating the following classes we will discuss three very important concepts of object oriented programming in Python: 

- Importance of properly documenting Python code and that relationship to help() and dir() functions in the command line
- Self keyword
- Encapsulation

Let us now start to construct some classes that will help perform the different analysis on a genome sequence that we have been programming during the workshop.


```python
import matplotlib.pyplot as plt     #used some methods in the derived class DNA_Sequence

#Base Class
class TextFile:
    """This class is meant for creating a DNA file analyzer.
    
    Attributes:
        text_file_sring (string) stores as a string the contents of the textfile        
    """
    
    def __init__(self, text_file_path = ""):         #initializaer/consructor method
        
        if TextFile.is_txt_file(text_file_path):
            self.text_file_string = TextFile.convert_file_to_string(text_file_path)
        else:
            TextFile.display_error_message() #could throw an exception instead but 
    
    
    
    #that's a more advanced subject we can cover later
    def convert_file_to_string(a_text_file_path):
        """Class function that opens a text file and stores its contents in a string.
        
        Args:
            file_name (string): name of a file to read from
        
        Returns:
            string: the content in the text file opened using the path entered as a parameter 
        """
        
        with open(str(a_text_file_path), "r") as a_text_file:
            a_text_file.readline()  #this command ignores the first lines(descriptions only) of the 
                                    #genome in the file 
            a_text_file_string = a_text_file.read()
        return a_text_file_string
    
    
    
    def display_error_message():
        """Class function that display's error message indicating that the entered path is has no .txt extension.        
        Args:
            None
            
        Returns: 
            None
        """
        
        print("ERROR MESSAGE. The path entered is not a path from a text file. Please check that the path entered is an available text file.\n")
    
    
    
    def is_txt_file(a_path):
        """Class function that identifies whether or not a path has a .txt extension or not. 
        
        Args: 
            a_path (string): path of a text file  
        Returns
            bool: true if file has .txt extension
        """
        
        paths_length = len(a_path)
        if paths_length > 4:
            extensions = a_path[paths_length - 4:]
            if extensions == ".txt":
              return True
            else: 
              return False 
        else:
            return False
```

This class is already ready to be tested! We will try to load 2 files. One the files will be a text file while the other one will not. We will know if the programm is working correctly if it opens and stores the content of the text file in a variable type string, and if it  displays and error message if the file is not a text file.

Notice that the methods in this class all have a string under the method's name that describes the purpose, arguments and returned values of the methods. This string is called a __docstring__. Docstrings are essential for other programmers to read and understand what you have coded. A good practice for programmers is to use docstrings and comments to properly document your code.

Now let us create a derived class that is specifically designed to hold a special type of text file: a DNA sequence. This class will have many method that will perform statistical analysis and visualizations of such analysis to help scientists acquire more insight about the composition of the DNA sequence that they are studying.


```python
#Derived class from the textfile class
class DNA_Sequence:                     #notice the sintax for inheritance
    """This class is meant for creating a DNA file analyzer
    
    """

    def __init__(self, a_path):
        text_file = TextFile(a_path)              #base class is used for initialization 
        self.genome = text_file.text_file_string  #not necessary but makes code clearer
    
    
    
    def get_genome(self):
        """Accesor of the class variable genome.

        """
        return self.genome
    
    
    
    def get_codons(self):
        """Return list with partition of the genome file. The returned list will have sublists of size 3 or
        less each.
        Args:
            None
        Returns: 
            list: list of codons, codons are lists of 3 nucleotides or less
        
        """
        strand_list = list(self.genome.replace("\n",""))  #text file's bottom whitespace is 
                                                          #eliminated in the str to list conversion
        strand_length = len(strand_list)
        
        residue_strand = strand_length % 3        #important : modulo operator
        strand_division = strand_length//3        #important : integer division

        strand_codons = []                        #alternatively strand_codons = list()
        temp_assignment_list = [0,0,0]
        for codon_num in range(strand_division):
            for k in range(3):
                temp_assignment_list[k] = strand_list[codon_num*3 + k]
            strand_codons.append(temp_assignment_list[:])
    
        if residue_strand != 0:
            residue_nucleotides_list = []
            for num in range(residue_strand):
                residue_nucleotides_list.append(strand_list[3*strand_division+num])
            strand_codons.append(residue_nucleotides_list)
            
        return strand_codons
        
        
        
    def concensus_sequence(self):
        """Mathod that finds the consensus sequence of a genome(i.e. sequence of most common nucleotide per columns). 
        
        Args:
            None
        Returns:
            list: contains de concensus sequence of the genome
        
        """
        genome_list = self.genome.strip().split("\n")     #strip() method here is important otherwise list would 
    
        #since lines in genome text files all have the same lenght 
        #we use range function with a any line in BRCA_list
        last_row = genome_list[-1]                              #saving last row apart from the BRCA_list
        lenght_last_row = len(last_row)                       #storing the length of the last row 
                                                              #before the loop for meory purposes

        consensus_sequence = list()                           #Could have used list[]
        for position in range(len(genome_list[0])):             #position of each letter in row the 70x70 part of the genome
            num_A, num_T, num_C, num_G = 0, 0, 0, 0           #Remember assignment in tuples
            for index in range(len(genome_list)-1):              
                if genome_list[index][position] == "A":
                    num_A += 1
                if genome_list[index][position] == "T":
                    num_T += 1
                if genome_list[index][position] == "C":
                    num_C += 1
                if genome_list[index][position] == "G":
                    num_G += 1
            if position < lenght_last_row:
                if last_row[position] == "A":
                    num_A += 1
                if last_row[position] == "T":
                    num_T += 1
                if last_row[position] == "C":
                    num_C += 1
                if last_row[position] == "G":
                    num_G += 1
            
            columns_most_frequent = max([(num_A, "A"),(num_T, "T"), (num_C, "C"), (num_G, "G")])
            consensus_sequence.append(columns_most_frequent[1])
        return consensus_sequence 
    
    
    
    def get_frequencies(self):
        """Method that finds the frequency each nucleotide in the genome class variable. 
        
        Args:
            None
        Returns:
            list: contains the total number of A's, C's, T's, and G's, in that specific order
        
        """
        nucleotides = ["A","C","T","G"]       #["Adenine","Citozine","Thyamine","Guanine"]
        nucleotide_count = []
        for nucleotide_letter in nucleotides:
            nucleotide_count.append(self.genome.count(nucleotide_letter))
        return nucleotide_count

    
    
    def bar_graph(self):
        """Atomatically generates and shows(using matplotlib.pyplot) a bar graph that ilustrates the frequency of
        the nucleotides in the genome file.
        
        Args:
            None
        Returns:
            None
        
        """
        plt.style.use('seaborn')
        plt.bar(["Adenine","Citozine","Thyamine","Guanine"], self.get_frequencies(), color = "#8233C5") #x axis list must be in the same
                                                                            # order as in getFrequency() method
        plt.title("Frequencies of Nucleotides")
        plt.xlabel("Nucleotides")
        plt.ylabel("Frequency")
        plt.show()
    
    
    
    def pie_chart(self):
        '''Atomatically generates and shows(using matplotlib) a Pie Chart that ilustrates the relative frequency
        of the nucleotides in the genome file.
        
        Args:
            None
        Returns:
            None
        
        '''
        nucleotide_names = ["Adenine","Citozine","Thyamine","Guanine"]
        colors = ["#8233C5", "#436AAC", "#33C58E", "#DE791E"]
        plt.pie(self.get_frequencies(),labels = nucleotide_names, colors = colors)
        plt.show()
    
    
    
    def complementary_strand(self):
        """
        
        Args:
            None
        Returns:
            str: contains nucleotides such that A's are changed by T's, G's are changed by C's and vice
            versa for both cases.

        """
        DNA_list = list(self.genome)
        for index in range(len(self.genome)):
            if DNA_list[index] == "A":
                DNA_list[index] = "T"
            elif DNA_list[index] == "T":
                DNA_list[index] = "A"
            elif DNA_list[index] == "G":
                DNA_list[index] = "C"
            elif DNA_list[index] == "C":
                DNA_list[index] = "G"
        return "".join(DNA_list).replace("\n","")
```

Every method in the DNA class has the word self as a parameter. The __self__ keyword is a special reserved keyword used mainly in classes, to indicate to the class that it must use the values associated to the instance that is invoking the methods or variables of the class.

Also, notice how the attributes and functions that manipulate them are binded together in such a way that both are safe from outside interference and misuse. In computer programming, whenever this happens in a class we say that the class provides __encapsulation__. Encapsulation organizes your code into a cleaner and more undestandable format and protects its attributes from bugs.   

In future examples, we shall see the roll that _accesors_ and _mutator_ play in __encapsulation__, but for now our examples don't require us to study this topic.
