Now we can count the number of aromatic aminoacids(i.e. phenylalanine (F),
tyrosine (Y), tryptophan (W)) in this protein. This can be achieved in several ways. 

By implementing a for loop and using all wa have learned we can code the following:


```python
lysozyme_protein_file = open("aromatics_protein_example.txt","r")

lysozyme_protein_file.readline()                            #reads line where the counter was set to

F_counter, Y_counter, W_counter = 0 , 0 , 0        #This is the same as declaring each variable separately

counters = [F_counter, Y_counter, W_counter]

for line in lysozyme_protein_file:
    F_counter += line.count("F")
    Y_counter += line.count("Y")
    W_counter += line.count("W")
    
lysozyme_protein_file.close()

print("""
        Number of phenylalanine(F's): {}
        Number of Tyrosine(Y's):      {}
        Number of Tryptophan(W's):    {}
""".format(F_counter, Y_counter, W_counter))
      
      
```

To avoid having to use a the close() method each time a file is opened, the keyword with is used. This keyword's syntax is for files as follows:

    with open('my_path/my_file.txt', 'r') as file_name:
        file_as_string = file_name.read()
The most important point to notice here is that manipulations, extractions or additions to a file must me done in the indented block enabeled by the with keyboard. Python automatically closes the file after the indented block is executed even if there happens to ocurre any error or exception along the way.

Let us look at the code we did in the previous code cell but this time implemented using the with keyword.
    

```python
import matplotlib.pyplot as plt      #imported now only ot be used for the following bar graph
                                     #the use of matplotlib will be further explained in future workshops
                                     #but keep in mind that import statements shoud go in the beginning of code

with open("aromatics_protein_example.txt","r") as lysozyme_protein_file:
    lysozyme_protein_file.readline()                    #reads line where the counter was set to

    F_counter, Y_counter, W_counter = 0 , 0 , 0        #This is the same as declaring each variable separately


    for line in lysozyme_protein_file:
        F_counter += line.count("F")
        Y_counter += line.count("Y")
        W_counter += line.count("W")

    print("""

        Number of Phenylalanine(F's): {}
        Number of Tyrosine(Y's):      {}
        Number of Tryptophan(W's):    {}

    """.format(F_counter,Y_counter,W_counter))
    
    print("The following is a bar graph ilustrating the previous information. This bar graph is implemented using matplotlib.pyplot. ")
    
    #Basic Bar graph in matplotlib 
    counters = [F_counter, Y_counter, W_counter]
    plt.style.use('seaborn')
    plt.bar(["Phenylalanine(F's)","Tyrosine(Y's)","Tryptophan(W's)"], counters, color = '#338844')
    plt.title("Aromatic aminoacids in Lysozyme")
    plt.show()
    
%time           # magic keywords will be discussed in future workshops
```

To analyze the genome in this file, it is important to remove the description on the top and to eliminate all the white space at the bottom of the file.

```python
BRCA_file = open("BRCA1_nucleotide.txt", "r")
BRCA_file.readline()         #this command ignores the first lines(descriptions only) of the genome in the file     
BRCA_file_data = BRCA_file.read()
  
BRCA_list = BRCA_file_data.strip().split("\n")   #strip() method here is important otherwise list would some have 
                                                 #null or "" element at the end

print(BRCA_list)

    
```

### How big is it? Lets find out.

```python
print("""
      Number of rows = {}
      Number of coulmns = {}
      """.format(len(BRCA_list),len(BRCA_list[0])))
```

There is still a mayor problem. The last row of the genome file is of different length than the other. This forces us to make a bipartition of the file(one partition containing this row and one without this row) and analyze this two partitions separately. 
We will see in the next code cell how to handle this problem.

```python
#since lines in genome text files all have the same lenght 
#we use range function with a any line in BRCA_list

last_row = BRCA_list[-1]                              #saving last row apart from the BRCA_list
lenght_last_row = len(last_row)                       #storing the length of the last row 
                                                      #before the loop for meory purposes

consensus_sequence = list()                           #Could have used list[]
for position in range(len(BRCA_list[0])):             #position of each letter in row the 70x70 part of the genome
    num_A, num_T, num_C, num_G = 0, 0, 0, 0           #Remember assignment in tuples
    for index in range(len(BRCA_list)-1):              
        if BRCA_list[index][position] == "A":
            num_A += 1
        if BRCA_list[index][position] == "T":
            num_T += 1
        if BRCA_list[index][position] == "C":
            num_C += 1
        if BRCA_list[index][position] == "G":
            num_G += 1
    if position < lenght_last_row:
        if last_row[position] == "A":
            num_A += 1
        if last_row[position] == "T":
            num_T += 1
        if last_row[position] == "C":
            num_C += 1
        if last_row[position] == "G":
            num_G += 1
            
    columns_most_frequent = max([(num_A, "A"),(num_T, "T"), (num_C, "C"), (num_G, "G")])
    consensus_sequence.append(columns_most_frequent[1])
    
    
    
print(consensus_sequence)            
            
    
```
